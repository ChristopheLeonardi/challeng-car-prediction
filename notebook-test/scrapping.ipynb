##  Récupération des liens des voitures

```python
def retrieve_links_cars():
    ua = UserAgent()

    # Liste pour stocker tous les href
    all_links = []

    # Page de départ
    i = 1

    # Continuer à scraper tant que nous avons moins de 1500 liens
    while len(all_links) < 1500:
        url = f'https://www.lacentrale.fr/listing?makesModelsCommercialNames=&options=&page={i}'

        headers = {'User-Agent': ua.random}
        response = requests.get(url, headers=headers)
        if response.status_code != 200:
            print(f"Request error for url page {url}, status code: {response.status_code}")
            i += 1
            continue
        soup = bs(response.content, 'html.parser')

        cars = soup.find_all('div', class_='searchCardContainer')
        print(f"No car found on the page {url}")

        for car in cars:
            if car is not None:  # Vérifiez que le car n'est pas None
                try:
                    href = car.find("a").get('href')
                    if href is not None:  # Vérifiez que href n'est pas None
                        # Ajoutez href à la liste all_links
                        all_links.append(href)
                    else:
                        print("Link has no href attribute")
                except AttributeError:  # attrape l'exception si car.find("a") est None
                    print("No <a> tag found in this element")
            else:
                print("No element found in specified class")

        # Ne pas oublier d'augmenter le compteur de la page
        i += 1

    # Retourne la liste all_links à la fin de la fonction
    return all_links
```

```python
def scrape_car_data(links):
    car_data_list = []

    # Spécifiez les headers de votre navigateur
    headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3"}

    for link in links:
        page = requests.get(link, headers=headers)
        soup = bs(page.content, 'html.parser')

        car_data = {}

        try:
            price = soup.find({"span"}, {"class":"PriceInformation_classifiedPrice__b-Jae"}).text
            price = price.replace("\u202f", "").replace("\xa0", "").replace("€", "").strip()
            price = int(price)
            car_data['Prix'] = price
        except Exception as e:
            print(f"An error occurred when extracting price data from page with the following link {link}: {e}")

        # Répétez le processus pour chaque information
        fields = ["year", "origin", "firstCirculationDate", "firstHand", "owners",
                  "mileage", "energy", "gearbox", "externalColor", "doors",
                  "ratedHorsePower", "powerDIN", "critAir", "co2", "marqueModele"]

        for field in fields:
            try:
                if field == "co2":
                    info = soup.find({"li"}, {"id":field}).find({"button"},{"class":"Hyperlink_Hyperlink_hyperlink Hyperlink_Hyperlink_secondary Hyperlink_Hyperlink_underline Co2_Co2_label ConsumptionSection_Co2Label__zTc6M"}).text
                elif field == "powerDIN":
                    info = soup.find({"li"}, {"id":field}).find({"span"},{"class":"PowerSection_noWrap___a0Nz"}).text
                elif field ==  "marqueModele":
                    info = soup.find({"div"}, {"id": "summary-information"}).find({"div"}, {"class":"Text_Text_text SummaryInformation_title__5CYhW Text_Text_headline3"}).text
                else:
                    info = soup.find({"li"}, {"id":field}).find({"span"},{"class":"Text_Text_text Text_Text_body1"}).text
                car_data[field] = info
            except Exception as e:
                print(f"An error has occurred when extracting {field} from page with the following link {link}: {e}")

        car_data_list.append(car_data)

    return car_data_list
```
