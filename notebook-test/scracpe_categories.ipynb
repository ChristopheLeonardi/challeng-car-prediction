```python
pattern = r'window.__PRELOADED_STATE_LISTING__ = ({.+})'
```

```python
match = re.fullmatch(pattern, text)
if match:
    preloaded_state = match.group(1)
    print(preloaded_state)
else:
    print("No match found.")
```

```python
raw_data = match.group(1)
raw_data
```
